```python
# Inicializa a variável 'soma' com zero. Esta variável armazenará o total dos números.
soma = 0

# Define uma lista de números que queremos somar.
numeros = [1,2,3]

# Itera sobre cada 'numero' na lista 'numeros' usando um laço for.
# A cada iteração, o valor do número atual é adicionado à 'soma'.
for numero in numeros:
  soma+=numero

# Imprime o valor final da 'soma' após percorrer todos os números da lista.
print(soma)
```

# # Códigos de números primos

```python
import math

def eh_primo(n):
    # Números menores que 2 não são primos. Por definição, primos são maiores ou iguais a 2.
    if n < 2:
        return False
    # O número 2 é o único número primo par.
    if n == 2:
        return True
    # Se o número for par (e maior que 2), ele não é primo.
    if n % 2 == 0:
        return False
    # Para otimizar, verificamos apenas divisores ímpares de 3 até a raiz quadrada de n.
    # Se um número tiver um divisor maior que sua raiz quadrada, ele também terá um menor.
    # O passo de 2 (range(..., 2)) garante que verificamos apenas números ímpares.
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        # Se encontrarmos qualquer divisor, o número não é primo.
        if n % i == 0:
            return False
    # Se nenhum divisor for encontrado, o número é primo.
    return True
```

# # Inversão de String

```python
def inverte_texto(texto):
  """Inverte uma string de texto, caractere por caractere."""
  # Inicializa uma string vazia que irá armazenar o texto invertido.
  invertido = ''
  # Itera sobre cada caractere (letra) na string de entrada 'texto'.
  for letra in texto:
    # Adiciona a letra atual ao *início* da string 'invertido'.
    # Isso efetivamente constrói a string invertida.
    invertido = letra + invertido
  # Retorna a string que foi construída de forma invertida.
  return invertido
# Solicita que o usuário digite um texto.
texto_original = input("Digite um texto para inverter: ")

# Chama a função 'inverte_texto' com o texto fornecido pelo usuário.
texto_invertido = inverte_texto(texto_original)

# Exibe o texto original e o texto invertido.
print(f"Texto original: {texto_original}")
print(f"Texto invertido: {texto_invertido}")
```
